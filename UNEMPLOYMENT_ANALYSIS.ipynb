# DATA ANALYSIS

### PAIR PLOT

```python
# to visualize the relationship between the employment and unemployment rates and the labour participation rate

# Reset index
df = df.reset_index()

# Define columns for pair plot
cols = ["unemployment_rate", "employed", "labour_participate_rate"]

print(">> to visualize the relationship between the employment and unemployment rates and the labour participation rate >>")
# Create pair plot
sns.pairplot(df[cols])
```

### PIE CHART

```python
# group data by region and calculate average unemployment rate

avg_unemployment = df.groupby('region')['unemployment_rate'].mean()

# create a pie chart
fig, ax = plt.subplots(figsize=(12, 12))
ax.pie(avg_unemployment, labels=avg_unemployment.index, autopct='%1.1f%%')
ax.set_title('Average Unemployment Rate by Region')

plt.show()
```

### SCATTER PLOT

```python
# analyze the trend of unemployment rate over time by plotting the data as a scatter plot

plt.rcParams["figure.figsize"] = (20, 10)
plt.scatter(df['date'], df['unemployment_rate'])
plt.title('Unemployment Rate by Date',size=30)
plt.xlabel('Date',size=30)
plt.ylabel('Unemployment Rate (%)',size=30)
print(">> analyze the trend of unemployment rate over time by plotting the data as a scatter plot >>")
plt.show()
```

### BAR PLOT

```python
# to show the unemployment rate for each region in the dataset

plt.rcParams["figure.figsize"] = (20, 10)
sns.barplot(x='region', y='unemployment_rate', data=df)
plt.title('Unemployment Rate by Region',size=20)
plt.xlabel('Region',size=20)
plt.ylabel('Unemployment Rate (%)',size=20)
# Get the bar plot
bar_plot = sns.barplot(x='region', y='unemployment_rate', data=df)
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=90) # Rotate x-tick labels by 90 degrees
print(">> to analyse the unemployment rate for each region in the dataset >>")
plt.show()
```

### BOX PLOT

```python
# to analyse the outliers

# Create the box plot
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(30, 20))
sns.boxplot(data=df, x="region", y="unemployment_rate", hue="frequency", ax=ax)
ax.set_title("Unemployment Rate by Region and Frequency",size=30)
ax.set_ylabel("Unemployment Rate",size=30)
ax.set_xlabel("\nRegion",size=35)
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
ax.set_xticklabels(ax.get_xticklabels(), fontsize=30)
plt.subplots_adjust(wspace=0.5)
print(">> to analyse the outliers >>")
plt.show()
```

### HEAT MAP

```python
# for finding the correlations

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
print(">> to analyse the correlations >>")
plt.show()
```

```python
# to visualizes the unemployment rate by region and date

# Pivot the data to get the unemployment rate by region and date
df_pivot = pd.pivot_table(df, values='unemployment_rate', index='region', columns='date', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(df_pivot, cmap='YlGnBu')
plt.title('Unemployment Rate by Region and Date',size=20)
plt.xlabel('Date',size=20)
plt.ylabel('Region',size=20)
print(">> to visualizes the unemployment rate by region and date >>")
plt.show()
```

```python
# Print the first 5 rows to verify the data was loaded correctly
print(df.head())
print("\n\n")

# Get some basic statistics on the unemployment rate column
print(df['unemployment_rate'].describe())
print("\n")

# Calculate the average unemployment rate for each region
region_avg_unemp = df.groupby('region')['unemployment_rate'].mean()
print("Avg unemployment in each region:\n\n",region_avg_unemp)
print("\n\n")

# Calculate the overall average unemployment rate for the entire dataset
overall_avg_unemp = df['unemployment_rate'].mean()
print("Overall average unemployment rate:", overall_avg_unemp)
print("\n\n")

# Calculate the average unemployment rate by year
df['year'] = pd.DatetimeIndex(df['date']).year
yearly_avg_unemp = df.groupby('year')['unemployment_rate'].mean()
print("yearly avg unemployment: \n",yearly_avg_unemp)
```
