```python
x1=df[['TV','Radio','Newspaper']]
y1=df[['Sales']]
print(x1)
print(y1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.5,random_state=1)
```

# MODEL CREATION

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

# MODEL FITTING

```python
print(x_train)
```

```python
print(y_train)
```

```python
x_test
```

```python
y_test
```

```python
x_train=x_train.astype(int)
y_train=y_train.astype(int)
x_test=x_test.astype(int)
y_test=y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaler=sc.fit_transform(x_train)
x_test_scaler=sc.fit_transform(x_test)
```

```python
model.fit(x_train_scaler,y_train)
```

```python
y_pred=model.predict(x_test_scaler)
```

# MODEL ACCURACY

```python
accuracy=model.score(x_test_scaler,y_test)
```

```python
plt.scatter(y_test,y_pred,color="red",s=12)
print(type(y_test))
y_test = y_test.ravel()
y_pred = y_pred.ravel()
slope, intercept = np.polyfit(y_test, y_pred, 1)
x = np.linspace(np.min(y_test), np.max(y_test), 100)
y = slope*x + intercept
plt.title("SALLES PREDICTION USING LINEAR REGRESSION")
plt.plot(y_test, slope*y_test + intercept, color='black')
plt.show()
```
